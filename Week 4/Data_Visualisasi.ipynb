```python
import pandas as pd
# Mengimpor pustaka yang diperlukan
import seaborn as sns
import matplotlib.pyplot as plt
```


prompt: beri kode untuk membaca dataset menjadi dataframe dengan df dari google drivepada google colab

```python
file_path = '/content/drive/MyDrive/Datasets/diamonds.csv'
df = pd.read_csv(file_path)
```

Beri kode untuk menampilkan jumlah unique value untuk masing-masing fitur(carat,cut,color,clarity,depth,table,price,x,y,z)

```python
# Menghitung jumlah nilai unik untuk setiap kolom
unique_counts = df[['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'price', 'x', 'y', 'z']].nunique()

# Menampilkan jumlah nilai unik untuk setiap kolom
print(unique_counts)
```

beri kode untuk mengelompokkan fitur yang merupakan numerical features(carat,depth,table,x,y,z) categorical features(cut,color,clarity), dan target(price)


```python
# Misalkan df adalah DataFrame Anda

# Menentukan fitur numerical
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Menentukan fitur categorical
categorical_features = ['cut', 'color', 'clarity']

# Menentukan target
target = ['price']

# Mengelompokkan fitur sesuai kategori
numerical_df = df[numerical_features]
categorical_df = df[categorical_features]
target_df = df[target]
```

beri kode untuk menampilkan bar chart dari fitur categorical


```python
# Misalkan categorical_df adalah DataFrame yang berisi fitur-fitur kategoris Anda

# Menggunakan seaborn untuk membuat bar chart untuk "cut"
plt.figure(figsize=(10, 6))  # Ukuran plot (opsional)
sns.countplot(data=categorical_df, x='cut', order=categorical_df['cut'].value_counts().index)

# Konfigurasi plot (opsional)
plt.title('Count of Cut Categories')
plt.xlabel('Cut')
plt.ylabel('Count')

# Menampilkan plot
plt.xticks(rotation=45)  # Untuk memutar label sumbu x jika diperlukan
plt.show()

# Menggunakan seaborn untuk membuat bar chart untuk "color"
plt.figure(figsize=(10, 6))
sns.countplot(data=categorical_df, x='color', order=categorical_df['color'].value_counts().index)
plt.title('Count of Color Categories')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Menggunakan seaborn untuk membuat bar chart untuk "clarity"
plt.figure(figsize=(10, 6))
sns.countplot(data=categorical_df, x='clarity', order=categorical_df['clarity'].value_counts().index)
plt.title('Count of Clarity Categories')
plt.xlabel('Clarity')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Mengimpor pustaka yang diperlukan
import matplotlib.pyplot as plt

# Misalkan categorical_df adalah DataFrame yang berisi fitur-fitur kategoris Anda

# Loop melalui setiap fitur kategoris dan membuat pie chart untuk masing-masing
categorical_features = ['cut', 'color', 'clarity']

for feature in categorical_features:
    # Menghitung jumlah pengamatan dalam setiap kategori
    category_counts = categorical_df[feature].value_counts()

    # Plot pie chart
    plt.figure(figsize=(8, 8))  # Ukuran plot (opsional)
    plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Pie Chart of {feature} Categories')

    # Menampilkan pie chart
    plt.axis('equal')  # Membuat pie chart berbentuk lingkaran
    plt.show()
```

```python
# Mengimpor pustaka yang diperlukan
import seaborn as sns
import matplotlib.pyplot as plt

# Misalkan numerical_df adalah DataFrame yang berisi fitur numerik Anda

# Loop melalui setiap fitur numerik dan membuat histogram untuk masing-masing
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

for feature in numerical_features:
    # Membuat histogram
    plt.figure(figsize=(8, 6))  # Ukuran plot (opsional)
    sns.histplot(numerical_df[feature], kde=True)  # Menggunakan kde=True untuk menambahkan kurva kepadatan
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)

    # Menampilkan histogram
    plt.show()
```

```python
# Mengimpor pustaka yang diperlukan
import matplotlib.pyplot as plt

# Misalkan numerical_df adalah DataFrame yang berisi fitur numerik Anda

# Fitur numerik
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Mengatur ukuran plot dan jumlah baris dan kolom
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Loop melalui setiap fitur numerik dan membuat histogram untuk masing-masing subplot
for i, feature in enumerate(numerical_features):
    row, col = divmod(i, 3)  # Menghitung baris dan kolom subplot yang sesuai
    ax = axes[row, col]

    # Membuat histogram
    ax.hist(numerical_df[feature], bins=30, edgecolor='k', color='blue')
    ax.set_title(f'Distribution of {feature}')
    ax.set_xlabel(feature)
    ax.set_ylabel('Frequency')

# Mengatur jarak antara subplot
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets

# Misalkan numerical_df adalah DataFrame yang berisi fitur numerik Anda

# Nama fitur numerik
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Fungsi untuk membuat histogram dengan kontrol bin width
def plot_histogram(bin_width):
    for feature in numerical_features:
        plt.figure(figsize=(8, 6))
        plt.hist(numerical_df[feature], bins=int((numerical_df[feature].max() - numerical_df[feature].min()) / bin_width), edgecolor='k')
        plt.title(f'Distribution of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Frequency')
        plt.show()

# Membuat kontrol interaktif
bin_width_slider = widgets.FloatSlider(
    value=0.1,
    min=0.01,
    max=2.0,
    step=0.01,
    description='Bin Width:',
    continuous_update=False,
)

# Menghubungkan fungsi dan kontrol interaktif
interact(plot_histogram, bin_width=bin_width_slider)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import ipywidgets as widgets

# Misalkan numerical_df adalah DataFrame yang berisi fitur numerik Anda
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Fungsi untuk membuat histogram dengan kontrol bin width
def plot_histogram(bin_width_carat, bin_width_depth, bin_width_table, bin_width_x, bin_width_y, bin_width_z):
    bin_widths = [bin_width_carat, bin_width_depth, bin_width_table, bin_width_x, bin_width_y, bin_width_z]
    for i, feature in enumerate(numerical_features):
        plt.figure(figsize=(8, 6))
        plt.hist(numerical_df[feature], bins=int((numerical_df[feature].max() - numerical_df[feature].min()) / bin_widths[i]), edgecolor='k')
        plt.title(f'Distribution of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Frequency')
        plt.show()

# Membuat kontrol interaktif untuk setiap fitur numerik
bin_width_sliders = [
    FloatSlider(value=0.1, min=0.01, max=2.0, step=0.01, description=f'Bin Width {feature}:', continuous_update=False)
    for feature in numerical_features
]

# Menghubungkan fungsi dan kontrol interaktif
interact(plot_histogram,
         bin_width_carat=bin_width_sliders[0],
         bin_width_depth=bin_width_sliders[1],
         bin_width_table=bin_width_sliders[2],
         bin_width_x=bin_width_sliders[3],
         bin_width_y=bin_width_sliders[4],
         bin_width_z=bin_width_sliders[5])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Misalkan numerical_df adalah DataFrame yang berisi fitur numerik Anda

# Nama fitur numerik
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Mengatur ukuran plot dan jumlah baris dan kolom
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Loop melalui setiap fitur numerik dan membuat boxplot untuk masing-masing
for i, feature in enumerate(numerical_features):
    row, col = divmod(i, 3)  # Menghitung baris dan kolom subplot yang sesuai
    ax = axes[row, col]

    # Membuat boxplot
    ax.boxplot(numerical_df[feature], vert=False)
    ax.set_title(f'Boxplot of {feature}')
    ax.set_xlabel(feature)

# Mengatur jarak antara subplot
plt.tight_layout()

# Menampilkan plot
plt.show()

```
